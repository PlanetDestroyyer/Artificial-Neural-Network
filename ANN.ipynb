```python
from sklearn.preprocessing import LabelEncoder
str_colms = ['Geography','Gender','Surname']

le = LabelEncoder()

for i in str_colms:
  df[i] = le.fit_transform(df[i])
```

```python
X = df.iloc[:,:-1]
```

```python
X.drop(columns=['Surname','CustomerId'],inplace=True)
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
df.head()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
se = StandardScaler()
```

```python
se.fit_transform(X)
```

```python
X = pd.DataFrame(se.fit_transform(X),columns=X.columns)
```

```python
X
```

```python
X.shape
```

```python
import tensorflow
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(11, input_dim=11, activation="relu"))
ann.add(Dense(7, activation="relu"))
ann.add(Dense(3, activation="relu"))
ann.add(Dense(1, activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
X_train.shape
```

```python
ann.fit(X_train,y_train,batch_size=100,epochs=100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_pred = ann.predict(X_test)
```

```python
print(accuracy_score(y_test,X_pred.round()*100))
```
